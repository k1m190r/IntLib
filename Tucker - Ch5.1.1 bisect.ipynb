# Bisect method for root finding

```python
roots = []
icount = 0


def bisect(f: Callable, x: IA, tol: float):
    global roots, icount
    icount += 1
    if f(x).has_zero:
        diam = x.rad * 2
        if diam < tol:
            roots.append(x)
        else:
            _x, x_ = x
            m = x.mid
            bisect(f, IA(_x, m), tol)
            bisect(f, IA(m, x_), tol)


def f(x):
    return sin(x) * (x - cos(x))


bisect(f, IA(-10, 10), 0.001)
icount, roots, len(roots)
```

```python
fig = figure(width=800, height=400)
x = np.linspace(-10, 10, 1000)
fig.line(x, f(x))
fig.add_layout(Span(location=0, dimension='width'))
for r in roots:
    _r, r_ = r
    fig.segment(x0=_r, y0=0, x1=r_, y1=0, color='red', line_width=10)
show(fig)
```
